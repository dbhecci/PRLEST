# Animacion

```python
fig = px.bar(drivers2021, x="points",
             y="Driver",
             color='Team',
             color_discrete_map=ColorTeams,
             animation_frame='round',
             height=800 )
fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
# Se muestra desde 1990

years_filter = circuit_races3['year']>1990
fig = px.scatter_geo(circuit_races3[years_filter], lat="lat", lon="lon", color="country",
                     hover_name="country", size="races",
                     animation_frame="year",
                     projection="natural earth")

fig.show()
```
